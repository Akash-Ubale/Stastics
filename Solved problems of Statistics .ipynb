### Problem Statement
A company used a specific brand of Tube Light in the past which has an average life of 1000 Hrs. 

A new brand has approached the company with new tube lights with same power at a lowe price. a sample of 120 tube lights were taken for testing, which yielded an average life of 1100 Hrs with SD of 90 Hrs. Should the company give the contract to this new company at 1% significance level ?

```python
# Given :

population_mean = 1000
sample_size = 120
sample_mean = 1100
standard_deviation = 90

# significance level
alpha = 0.01
```

Now stating rthe Hypothesis :

Null Hypothesis $H_0$ = average life >= 1000

Aalternate Hypothesis $H_1$ = average life < 1000

So it's an one tailed test

the significance level is 1% means for critical region the area under the distribution curve is 0.01.

Now let's find the Z-value from the Z-table for significance level = 0.01

As exact 0.01 is not present we will select the closest value to 0.01, which we found as 0.009.

hence the Z-value (z_t) = -2.33

Now  the critical region < -2.33 



```python
z_t = -2.33
```

Now let's calculate the Z-value for our problem.

$$ z = \frac{Sample  Mean - Population  Mean}{Standard  Error} $$

$$ Standard Error = \frac{Standard  Deviation}{\sqrt{Sample  Size}}$$

```python
# calculating standard error.
import math

standard_error = standard_deviation / (math.sqrt(sample_size))
standard_error
```

```python
# calculating z-value
z_c = (sample_mean - population_mean)/standard_error
z_c
```

if the z_c is lesser than the z_t then Rejecting the Null Hypothesis.


```python
if z_c < z_t:
    print("Rejecting the Null Hypothesis, and the contract should be given to the Old company.")
else:
    print("Accepting the Null Hypothesis, and the contract should be given to the New Company.")
```

# Problem Statement:

The means of two samples of size 10 and 8, from two normal populations is 210.40 and 208.92 respectively. The sum of squares of deviations from their means is 26.94 and 24.50 respectively.
Assuming population with equal variances can we consider the normal populations have equal means? (significance level = 5%)

```python
# Given :

size_s1 = 10
size_s2 = 8
mean_s1 = 210.40
mean_s2 = 208.92
sum_sq_devi_s1 = 26.94
sum_sq_devi_s2 = 24.50
```

Null Hypothesis $H_0$ : The normal populations have **equal means. i.e. mean_pop1 = mean_pop2** 

Alternate Hypothesis $H_1$ : The normal populations have **unequal means. i.e. mean_pop1 $\neq$ mean_pop2**

Hence it's a two tailed test.

As the sample size is very small , we are using T-test.

So Find t-value from the T-table for :
    1. degree of freedom = size_s1 + size_s2 - No. of samples 
                         = 10 + 8 - 2 = 16
    2. Two tailed test with alpha = 0.05
    
 

so our t-value from the table is 2.120

```python
t_t = 2.120
```

This is the threshold t-value

Now if our calculated t-value lies in the critical region we will reject the Null Hypothesis $H_0$.

Calculating the T-value :

$$ t_c = \frac{mean_{S1} - mean_{S2}}{Standard Error} $$

$$ Standard Error = \frac{sd_{S1} + sd_{S2}}{\sqrt{size_{s1} + size_{s2} - 2}} $$

$ sd = \sqrt{Variance}$

$ variance = \frac{1}{n} \sum (x_i - \bar x)  $

```python
import math

# For S1
sum_sq_devi_s1 = 26.94
var_s1 = sum_sq_devi_s1/size_s1
sd_s1  = math.sqrt(var_s1)
print("sd_s1: ", sd_s1)

# For S2
var_s2 = sum_sq_devi_s2/size_s2
sd_s2  = math.sqrt(var_s2)
print("sd_s2: ", sd_s2)

# Now calculating the standard error.
sd_error = (sd_s1 + sd_s2) / math.sqrt(size_s1 + size_s2 - 2)
print("sd_error:",sd_error)

# now calculating t-value for our problem.
t_c = (mean_s1 - mean_s2) / sd_error
print("t_c:", t_c)
```

Let's see where this value lies on our normal distribution.



t_c doesn't lies in the critical region.

hence accepting the null hypothesis $H_0$ : The normal populations have **equal means. i.e. mean_pop1 = mean_pop2**
